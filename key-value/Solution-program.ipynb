```python
import pandas as pd
import time
```

```python
data=pd.read_csv("input.csv",header=None)   #taking data from input file
```

```python
data.shape
```

```python
heading=['key','value'] #to add column-headings to our data
```

```python
data.columns=heading
```

```python
print(data)  
```

```python
dictionary={}    #i am using a dictionary in python, which is similar to hashmaps in c++
```

```python
for i in range(len(data['key'])):    #here it maps , each of key from column named key with its value
    #from column named value
    dictionary[data['key'][i]]=data['value'][i]
```
