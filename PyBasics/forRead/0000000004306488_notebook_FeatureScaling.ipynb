```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tts
from matplotlib import pyplot as plt
```

```python
# Adding more features to dataset

boston = datasets.load_boston()
df = pd.DataFrame(boston.data)
a = boston.data.shape[1]
b=a
for i in range(a):
    for j in  range(i,a):
        df[b] = df[i]*df[j]
        b += 1
    
# new data
new_x = df.values
y = boston.target
alg1 = LinearRegression()
x_train_new, x_test_new, y_train, y_test = tts(new_x,y,test_size=0.2,random_state=1)
alg1.fit(x_train_new,y_train)
new_score = alg1.score(x_test_new,y_test)
y_pred_new = alg1.predict(x_test_new)

# old data
old_x = boston.data
alg2 = LinearRegression()
x_train_old, x_test_old, y_train, y_test = tts(old_x,y,test_size=0.2,random_state=1)
alg2.fit(x_train_old,y_train)
old_score = alg2.score(x_test_old,y_test)
y_pred_old = alg2.predict(x_test_old)

print("Old Score: ",old_score)
print("New Score: ",new_score)


```

```python
# parabolic Equation

def y_approx(x):
    random_val = np.random.random(x.shape)*1000-500
    return 3*x*x + 5*x + 6 + random_val   

x = np.array([[i] for i in range(-50,50,1)])
y = y_approx(x)
clf = LinearRegression()
clf.fit(x,y)
y_pred = clf.predict(x)
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.show()
```

```python
x1 = np.c_[x,x**2]
clf2 = LinearRegression()
clf2.fit(x1,y)
y1_pred = clf2.predict(x1)
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.plot(x,y1_pred)
plt.show()
```
