```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
boston=np.genfromtxt('project1.csv',dtype=None,delimiter=',')
df=pd.DataFrame(boston)
y=df[13]
del df[13]
```

```python
count=13
for i in range(13):
    for j in range(i,13):
        df[count]=df[i]*df[j]
        count+=1
```

```python
df[count]=np.ones(df.shape[0])
x=np.array(df)
```

```python
rows=x.shape[0]
cols=x.shape[1]
rows,cols,y.shape
```

```python
def cost(x,y,m):
    total=0
    for i in range(rows):
        total=total+((y[i]-(m*x[i]).sum())**2)/rows
    return total
```

```python
def stochastic_step_gradient(x,y,l_rate,m):
    slope_m=np.zeros(cols)
    for i in range(rows):
        tempx=x[i]
        tempy=y[i]
        for j in range(cols):
            #print(x[i,j])
            slope_m[j]=(2/rows)*tempx[j]*((m*tempx).sum()-tempy)
            m[j]=m[j]-l_rate*slope_m[j]
    return m
```

```python
def mini_batch_step_gradient(x,y,l_rate,m):
    slope_m=np.zeros(cols)
    for i in range(rows//10):
        tempx=x[i]
        tempy=y[i]
        for k in range(20):
            for j in range(cols):
                slope_m[j]=(2/rows)*tempx[j]*((m*tempx).sum()-tempy)
        for j in range(cols):
            m[j]=m[j]-l_rate*slope_m[j]
    return m
```

```python
def step_gradient(x,y,l_rate,m):
    slope_m=np.zeros(cols)
    for i in range(rows):
        tempx=x[i]
        tempy=y[i]
        for j in range(cols):
            #print(x[i,j])
            slope_m[j]+=(2/rows)*tempx[j]*((m*tempx).sum()-tempy)
    for i in range(cols):
        m[i]=m[i]-l_rate*slope_m[i]
    return m
```

```python
def gradient_descent(x,y,l_rate,n_itr):
    m=np.zeros(cols)
    print("intial: ",cost(x,y,m))
    for i in range(n_itr):
        m=stochastic_step_gradient(x,y,l_rate,m)
        print(i," : ",cost(x,y,m))
    return m
```

```python
def run(x,y):
    l_rate=0.04
    n_itr=270
    m=gradient_descent(x,y,l_rate,n_itr)
    return m
```

```python
m=run(x,y)
m
```
