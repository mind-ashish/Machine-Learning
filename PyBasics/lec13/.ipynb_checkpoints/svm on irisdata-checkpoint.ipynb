```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
X=iris.data[:,0:2]    #just like logistic regression ,in SVm we work on two features, so take first 2 cols out of 4.
Y=iris.target
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)       #plotting col 0 and 1 , based on class they belong two. Since 3 types of folower.
                                      #3 diff colors
plt.xlabel('sepal length') #xlabel gives label to x axis.,y label to y axis.
plt.ylabel('sepal width')
plt.show()
```

```python
#similarly plotting last 2 cols.
X=iris.data[:,2:]                        #picking col from 2 to end ,means last 2 out of 4. 
Y=iris.target
plt.scatter(X[:,0],X[:,1],c=Y)
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```

```python
#1. svm with linear kernel. means no higher degree of space added. Just linear decision boundaries. Similarity increase,decrease linearly
from sklearn import svm
```

```python
X=iris.data[:,0:2]   #pick first two cols.
Y=iris.target
clf=svm.SVC(kernel='linear',C=1.0)   #C=1.0 by default also, it gives regularisation.
```

```python
clf.fit(X,Y)    #fit with 2 features and target, i.e first 2 cols and Y.
```

```python
x1=X[:,0]    #first col
```

```python
x2=X[:,1]     #second col
```

```python
h=0.02
x1_min,x1_max = x1.min() - 1, x1.max() + 1
x2_min,x2_max = x2.min() - 1, x2.max() + 1
a = np.arange(x1_min, x1_max, h)
b = np.arange(x2_min, x2_max, h)
xx, yy = np.meshgrid(a, b)
```

```python
plot_data=np.c_[xx.ravel(),yy.ravel()]    #taking all pts nearby x1min x1max,x2min,x2max in form of array
```

```python
plot_data
```

```python
y_pred=clf.predict(plot_data)      #predicting class for these pts ,using already fit features.
```

```python
plt.scatter(xx.ravel(),yy.ravel(),c=y_pred)   #plotting these pts ,with respect to their predicted class
plt.show()
```

```python
plt.title('SVC with linear kernel')   #to give title on top
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.xticks(())             #xticks, is used to give label to points, pass nothing in xticks here
plt.yticks(())

plt.scatter(xx.ravel(),yy.ravel(),c=y_pred,cmap=plt.cm.coolwarm)   #scattering plts with respect to class predicted as done above
#cmap give contrasting color to points wrt background
plt.scatter(X[:,0],X[:,1],c=iris.target)     #also scattering col1 ,col2 pts with respect to class they belong

plt.show()
```

```python
#2. Linear SVC
clf=svm.LinearSVC(C=1.0)
clf.fit(X,Y)
```

```python
h=0.02
x1_min,x1_max = x1.min() - 1, x1.max() + 1
x2_min,x2_max = x2.min() - 1, x2.max() + 1
a = np.arange(x1_min, x1_max, h)
b = np.arange(x2_min, x2_max, h)
xx, yy = np.meshgrid(a, b)
plot_data=np.c_[xx.ravel(),yy.ravel()]   
y_pred=clf.predict(plot_data)
```

```python
plt.title('Linear SVC')   
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.xticks(())             
plt.yticks(())

plt.scatter(xx.ravel(),yy.ravel(),c=y_pred,cmap=plt.cm.coolwarm)    
plt.scatter(X[:,0],X[:,1],c=iris.target)

plt.show()
```

```python
#3.Svc with RBF kernel(basically gaussian)
```

```python
clf=svm.SVC(kernel='rbf',gamma=0.7,C=1.0)   #gamma is 1/sigma square. So gamma increase, then steep changes(decrement/increment) in similarity. in similarity
clf.fit(X,Y)                                #low value of C means regularity is present.
```

```python
h=0.02
x1_min,x1_max = x1.min() - 1, x1.max() + 1
x2_min,x2_max = x2.min() - 1, x2.max() + 1
a = np.arange(x1_min, x1_max, h)
b = np.arange(x2_min, x2_max, h)
xx, yy = np.meshgrid(a, b)
plot_data=np.c_[xx.ravel(),yy.ravel()]   
y_pred=clf.predict(plot_data)
```

```python
plt.title('Gaussian SVM or SVC with rbf kernel')   
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.xticks(())             
plt.yticks(())

plt.scatter(xx.ravel(),yy.ravel(),c=y_pred,cmap=plt.cm.coolwarm)    
plt.scatter(X[:,0],X[:,1],c=iris.target)

plt.show()
```

```python
#4.Polynomial kernel
#unlike logistic classifier where we need to add feature sq and cube in linear boundary itself, here we can do directly
```

```python
clf=svm.SVC(kernel='poly',degree=3,C=1.0)   #degree is degree of poly
clf.fit(X,Y)
```

```python
h=0.02
x1_min,x1_max = x1.min() - 1, x1.max() + 1
x2_min,x2_max = x2.min() - 1, x2.max() + 1
a = np.arange(x1_min, x1_max, h)
b = np.arange(x2_min, x2_max, h)
xx, yy = np.meshgrid(a, b)
plot_data=np.c_[xx.ravel(),yy.ravel()]   
y_pred=clf.predict(plot_data)
```

```python
plt.title('SVC with polynomial kernel')   
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.xticks(())             
plt.yticks(())

plt.scatter(xx.ravel(),yy.ravel(),c=y_pred,cmap=plt.cm.coolwarm)    
plt.scatter(X[:,0],X[:,1],c=iris.target)

plt.show()
```
