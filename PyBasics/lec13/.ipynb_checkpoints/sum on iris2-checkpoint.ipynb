```python
#here we are trying to explore svm on iris dataset with different type of kernel,varying C , gamma and degree.
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
import numpy as np
```

```python
print(iris.DESCR)
```

```python
X=iris.data[:,2:4]
Y=iris.target
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.title('Plot')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```

```python
from sklearn import svm
```

```python
def plotdata(x1,x2):
    h=0.02
    x1_min,x1_max = x1.min() - 1, x1.max() + 1
    x2_min,x2_max = x2.min() - 1, x2.max() + 1
    a = np.arange(x1_min, x1_max, h)
    b = np.arange(x2_min, x2_max, h)
    xx, yy = np.meshgrid(a, b)
    plot_data=np.c_[xx.ravel(),yy.ravel()] 
    return plot_data,xx,yy
```

```python
def plots(ch,X,Y,C,gamma=0.7,degree=1):   #ch is choice for  kernel
    if ch==1:
        for i in range(1,C,4999):
            print('C=',i)
            clf=svm.SVC(kernel='linear',C=i)
            clf.fit(X,Y)
            plot_data,xx,yy=plotdata(X[:,0],X[:,1])
            y_pred=clf.predict(plot_data)
            
            plt.title('SVC with linear kernel')   
            plt.xlabel('petal length')
            plt.ylabel('petal width')
            plt.xticks(())             
            plt.yticks(())

            plt.scatter(xx.ravel(),yy.ravel(),c=y_pred,cmap=plt.cm.coolwarm)    
            plt.scatter(X[:,0],X[:,1],c=iris.target)

            plt.show()
    if ch==2:
        for i in range(1,C,4999):
            print('C=',i)
            clf=svm.LinearSVC(C=i)
            clf.fit(X,Y)
            plot_data,xx,yy=plotdata(X[:,0],X[:,1])
            y_pred=clf.predict(plot_data)
            
            plt.title('Linear SVC')   
            plt.xlabel('petal length')
            plt.ylabel('petal width')
            plt.xticks(())             
            plt.yticks(())

            plt.scatter(xx.ravel(),yy.ravel(),c=y_pred,cmap=plt.cm.coolwarm)    
            plt.scatter(X[:,0],X[:,1],c=iris.target)

            plt.show()
    if ch==3:
        for i in range(1,C,4999):
            j=0.5
            while(j<gamma):
                print('C=',i,"",'gamma=',j)
                clf=svm.SVC(kernel='rbf',gamma=j,C=i)   #gamma is 1/sigma square. So gamma increase, then steep changes(decrement/increment) in similarity. in similarity
                clf.fit(X,Y)
                plot_data,xx,yy=plotdata(X[:,0],X[:,1])
                y_pred=clf.predict(plot_data)
                
                plt.title('Gaussian SVM or SVC with rbf kernel')   
                plt.xlabel('petal length')
                plt.ylabel('petal width')
                plt.xticks(())             
                plt.yticks(())

                plt.scatter(xx.ravel(),yy.ravel(),c=y_pred,cmap=plt.cm.coolwarm)    
                plt.scatter(X[:,0],X[:,1],c=iris.target)

                plt.show()
                j+=2
    if ch==4:
        for i in range(1,C,4999):
            for j in range(1,degree,1):
                print('C=',i,"",'degree=',j)
                clf=svm.SVC(kernel='poly',degree=j,C=i)   #degree is degree of poly
                clf.fit(X,Y)
                plot_data,xx,yy=plotdata(X[:,0],X[:,1])
                y_pred=clf.predict(plot_data)
                
                plt.title('SVC with polynomial kernel')   
                plt.xlabel('petal length')
                plt.ylabel('petal width')
                plt.xticks(())             
                plt.yticks(())

                plt.scatter(xx.ravel(),yy.ravel(),c=y_pred,cmap=plt.cm.coolwarm)    
                plt.scatter(X[:,0],X[:,1],c=iris.target)

                plt.show()
            
```

```python
#svc with linear kernel
plots(1,X,Y,20000)  
```

```python
#svc with linear svc
plots(2,X,Y,20000)
```

```python
#svc with kbf kernel
plots(3,X,Y,20000,8)
```

```python
#svc with polynomial kernel
plots(4,X,Y,20000,8,6)
```
