```python
from sklearn import datasets
```

```python
import pandas as pd
```

```python
iris=datasets.load_iris()
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
#email
```

```python
for i in range(0,150,1):
    if df.loc[i]['sepal length (cm)']<5.100000:
        df.loc[i]['sepal length (cm)']=0
    if df.loc[i]['sepal length (cm)']<5.800000:
        df.loc[i,'sepal length (cm)']=1
    if df.loc[i]['sepal length (cm)']<6.400000:
        df.loc[i]['sepal length (cm)']=2
    else:
        df.loc[i]['sepal length (cm)']=3
```

```python
df['sepal length (cm)'].unique()
```

```python
for i in range(0,150,1):
    if df.loc[i]['sepal width (cm)']<2.800000:
        df.loc[i]['sepal width (cm)']=0
    elif df.loc[i]['sepal width (cm)']<3.000000:
        df.loc[i]['sepal width (cm)']=1
    elif df.loc[i]['sepal width (cm)']<3.300000:
        df.loc[i]['sepal width (cm)']=2
    else:
        df.loc[i]['sepal width (cm)']=3
```

```python
df['sepal width (cm)'].unique()
```

```python
for i in range(0,150,1):
    if df.loc[i]['petal length (cm)']<1.600000:
        df.loc[i]['petal length (cm)']=0
    elif df.loc[i]['petal length (cm)']<4.350000:
        df.loc[i]['petal length (cm)']=1
    elif df.loc[i]['petal length (cm)']<5.100000:
        df.loc[i]['petal length (cm)']=2
    else:
        df.loc[i]['petal length (cm)']=3
```

```python
df['petal length (cm)'].unique()
```

```python
for i in range(0,150,1):
    if df.loc[i]['petal width (cm)']<0.300000:
        df.loc[i]['petal width (cm)']=0
    elif df.loc[i]['petal width (cm)']<1.300000:
        df.loc[i]['petal width (cm)']=1
    elif df.loc[i]['petal width (cm)']<1.800000:
        df.loc[i]['petal width (cm)']=2
    else:
        df.loc[i,'petal width (cm)']=3
```

```python
df['petal width (cm)'].unique()
```
