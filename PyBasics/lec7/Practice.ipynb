```python
 #this all is be done using makegrid fn.

def makegrid(x1, x2, h = 0.02):
    x1_min,x1_max = x1.min() - 1, x1.max() + 1
    x2_min,x2_max = x2.min() - 1, x2.max() + 1
    a = np.arange(x1_min, x1_max, h)
    b = np.arange(x2_min, x2_max, h)
    xx, yy = np.meshgrid(a, b)
    return xx, yy
```
