```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
pp=pd.read_csv("0000000000002419_training_ccpp_x_y_train.csv")
pp.shape
```

```python
pp
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_train=pp.iloc[:,0:4]
```

```python
x_train.shape
```

```python
y_train=pp.iloc[:,4:5]
y_train.shape
```

```python
scaler=StandardScaler()
```

```python
clf=LinearRegression()
```

```python
x_trainScaled=scaler.fit_transform(x_train)
```

```python
x_test=pd.read_csv("0000000000002419_test_ccpp_x_test.csv")
```

```python
x_testSCaled=scaler.transform(x_test)
```

```python
clf.fit(x_trainScaled,y_train)
```

```python
yp=clf.predict(x_testSCaled)
```

```python
yp.shape
```

```python
yp
```

```python
df=pd.DataFrame(yp)
```

```python
df.to_csv("Solution.csv",index=False,encoding='utf-8')
```
