```python
from sklearn import datasets
boston=datasets.load_boston()
import pandas as pd
df=pd.DataFrame(boston.data)
print(boston.data.shape)
i,count=0,13
#to add new col :df[col]=df[oldcol]
#to add new feature
while i<=12:
    j=i
    while j<=12:
        df[count]= df[i]*df[j]
        j+=1
        count+=1
    i+=1
df
#scaling req in gradient descent only, not in linear regression
from sklearn.linear_model import LinearRegression
alg=LinearRegression()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,boston.target,random_state=1)
alg.fit(x_train,y_train)
alg.score(x_test,y_test)
```

```python

import numpy as np
def y_approx(x):
    random_val=np.random.random(x.shape)*1000-500   #x.shape will return y in 100*1
    return 3*x*x +5*x+6+random_val
#np.random.random.random(100)*100 #random 100 value b/w 0 to 1. *100 to get b/w 1 to 100
#np.random.random.random(100)*100-50 #to shift mean by 50

#x=[[i] for i in range(-50,50,1)]

x=np.array([[i] for i in range(-50,50,1)])
#y=3*x*x + 5*x+6
y=y_approx(x)



from matplotlib import pyplot as plt



plt.scatter(x,y)
plt.show()

```

```python
from sklearn.linear_model import LinearRegression
alg=LinearRegression()
alg.fit(x,y)
```

```python

xt=np.random.random(x.shape)*100-25
yp=alg.predict(xt)
```

```python
plt.plot(xt,yp)
plt.scatter(x,y)
plt.show()
```
