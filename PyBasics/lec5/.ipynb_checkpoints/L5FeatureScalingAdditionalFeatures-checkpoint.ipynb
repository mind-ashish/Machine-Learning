```python
from sklearn import preprocessing
```

```python
scaled_data = preprocessing.scale(boston.data)
df2 = pd.DataFrame(scaled_data)
df2.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_data, boston.target, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
def y_approx(x):
    random_val = np.random.random(x.shape) * 1000 - 500
    return 3 * x * x + 5 * x + 6 + random_val
```

```python
import numpy as np
x = np.array([[i] for i in range(-50, 50, 1)])
y = y_approx(x)
```

```python
y
```

```python
from matplotlib import pyplot as plt
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearRegression()
clf.fit(x, y)
y_pred_1 = clf.predict(x)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred_1)
plt.show()
```

```python
x_new = np.c_[x, x**2]
clf2 = LinearRegression()
clf2.fit(x_new, y)
y_pred_2 = clf2.predict(x_new)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred_1)
plt.plot(x, y_pred_2)
plt.show()
```
