```python
from sklearn import datasets
```

```python
boston=datasets.load_boston()
boston.data
```

```python
import pandas as pd
bd= pd.DataFrame(boston.data)  #to view boston data
```

```python
bd   
```

```python
#col 11 has too large values, col 0 has small. So scale
#to import scaling fns
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(boston.data,boston.target,random_state=1)
#this has splitted data into training and test data with x var & y var as usual.
#now we need to scale the x vars in both xtrain & xtest by same factor before fitting & testing by score.
```

```python
x_train_scaled=scaler.fit_transform(x_train)  #this is scaling x_train & storing scale factor
df=pd.DataFrame(x_train_scaled)
df  #scaled data
```

```python
from sklearn.linear_model import LinearRegression
alg=LinearRegression()
alg.fit(x_train_scaled,y_train)
```

```python
#fitting is done, now before checking score, we need to scale x_test also by same scale factor
```

```python
x_test_scaled=scaler.transform(x_test)
```

```python
#now testing
alg.score(x_test_scaled,y_test)
```

```python
#taking y predict
yp=alg.predict(x_train_scaled)
```
