```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([1,3,5,7,9,7,4])
y1=np.array([-5,3,-11,6,3,-2.5,0])

plt.scatter(x1,y1)
plt.show()
```

```python
plt.plot(x1,y1,color='red')   #to plot line
plt.show()
```

```python
x2=np.array([2,3,8])
y2=np.array([2,3,8])
plt.scatter(x2,y2)    #scatter locate pts on a graph. Takes arrays:x,y,z etc of co-ordinate of pts as input
plt.show()            #show fn is used to print this graph
```

```python
x2=np.array([2,3,8])
y2=np.array([1,4,2])
z2=np.array([3,5,5])
plt.scatter(x2,y2,z2)    
plt.show()            
```

```python
plt.plot(x2,y2,z2,color='black')  #since 3d
plt.show()
```

```python
##linear regression
```

```python
from sklearn.linear_model import LinearRegression #to generate best line
from sklearn.model_selection import train_test_split   #to split given data into train & test data
from sklearn import datasets       #to use datasets
```

```python
alg=LinearRegression()
#training 
alg.fit(x1.reshape(7,1),y1)   #reshape fn changed it to corresponding 2d array of 7*1 since fit require 2d array of x points/input
```

```python
##testing
xp=[[2],[4],[6]]  #to predict against given value of x. #to plot best line
yp=alg.predict(xp)
plt.plot(xp,yp)
plt.show()
```

```python
plt.scatter(x1,y1)     #to scatter pts and see best line
plt.plot(xp,yp)
plt.show()
```

```python
plt.plot(x1,y1,'g^')  #to plot x1,y1 pts in form of boxes.
plt.plot(xp,yp,marker="*",color='grey')  #to show xp,yp pts on best line.
plt.show()
```

```python
alg.coef_ , alg.intercept_
```

```python
bc= datasets.load_breast_cancer()   #to load a dataset
```

```python
bc
```

```python
bc.DESCR    #to access DESCR field from file
```

```python
bc.data.shape
```

```python
bc.target.shape   #means 569*30 order matrix for input var Xi & 569*1 matrix for output Yi.
```

```python
#we split this data bc into 2 parts training data with x & y, and testing data with x & y, by using train test split, which take
#bc.data & bc.target as input & return xtrain,xtest,ytrain,ytest
x_train,x_test,y_train,y_test=train_test_split(bc.data,bc.target)
```

```python
#using the training data to train
alg_bc=LinearRegression()
```

```python
alg_bc.fit(x_train,y_train)
```

```python
#to generate y predict against testing data xtest
yp=alg_bc.predict(x_test)
plt.plot(x_train,y_train,'g^')
plt.plot(x_test,yp)
plt.show()
```

```python
#to see prediction accuracy we use score fn
alg_bc.score(x_test,y_test)   #it takes x test and y test , create y predict & compare it with ytest to give score.
```
