```python
from sklearn.linear_model import LinearRegression
alg_boston=LinearRegression()
alg_boston.fit(boston.data,boston.target)
```

```python
from sklearn.model_selection import train_test_split
x_train,xtest,y_train,ytest= train_test_split(boston.data,boston.target,test_size=0.2,random_state=1)  #for same result use random state
x_train.shape,xtest.shape,y_train.shape,ytest.shape
```

```python
alg_boston.fit(x_train,y_train)
```

```python
y_pred=alg_boston.predict(xtest)
alg_boston.score(xtest,ytest)   #find ypredict & compare it with ytest ,to check how ideally linear regression works.1->good, 0->poor

#overfit:huge gap b/w training & testing data &underfit
alg_boston.score(x_train,y_train)
```
