```python
import pandas as pd
diabetes=pd.read_csv("training_diabetes_x_y_train (1).csv")
x_train=diabetes.iloc[:,0:10]    #since col 0 to 9 contain input, use it as x_train
y_train=diabetes.iloc[:,10:11]    #last col represent output/Y it is y tain
from sklearn.linear_model import LinearRegression
diabetes_lr=LinearRegression()
diabetes_lr.fit(x_train,y_train)

#test data
diabetes_test=pd.read_csv("test_diabetes_x_test.csv")
xp=diabetes_test.iloc[:,0:10]
import numpy as np
yp=diabetes_lr.predict(xp)
yp=np.around(yp,decimals=5)    #yp is np array, to round off to 5 decimal places.

df=pd.DataFrame(yp)
df.to_csv("Solution.csv",index=False,encoding='utf-8')
ans=pd.read_csv("Solution.csv")
ans
```
