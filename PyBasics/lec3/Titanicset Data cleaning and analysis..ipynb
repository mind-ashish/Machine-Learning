```python
#we hv two to load data: using pandas(using path/or saved file) or using Datasets of sklearn(inbuilt dataset).
#use dataframe to load dataset of sklearn in pandas form.

import pandas as pd
titanic_original=pd.read_csv("train.csv")

#it is a traing data. with target as survived column & rest columns are data columns.

titanic=titanic_original.copy()
titanic
```

```python
#we need to 
#1.remove unwanted cols
#2.if possible convert string col to int for analysis
#3.replace where age is not given by mean age of survivor, if missing age is of survivor & vice versa
#4.plot graph.
```

```python
#deleting unnecessarys cols
titanic.drop('PassengerId',axis=1,inplace=True)
titanic.drop('Name',axis=1,inplace=True)
titanic.drop('Ticket',axis=1,inplace=True)
titanic.drop('Cabin',axis=1,inplace=True)
```

```python
titanic
```

```python
titanic.at[1,'Sex']    #to access ith row, jth col with  row name & col name use titanic.at['rowname','col name'] 
# or use titanic.at[i,j] or titanic.at[i,'colname'] etc.
```

```python
#2. changing strings to int. if in sex column, if entry is male , then replace by 0, else by 1.
for i in range(0,891,1):
    if titanic.at[i,'Sex']=='male':
        titanic.at[i,'Sex']=0
    else:
        titanic.at[i,'Sex']=1
```

```python
titanic
```

```python
#2. In embarked col if entry ='S', replace by 1, 'C' replace by 0, 'Q' replace by 2.
for i in range(0,891,1):
    if titanic.at[i,'Embarked']=='C':
        titanic.at[i,'Embarked']=0
    elif titanic.at[i,'Embarked']=='S':
        titanic.at[i,'Embarked']=1
    else:
        titanic.at[i,'Embarked']=2
```

```python
titanic
```

```python
#3.replacing the missing age of unsurvived by mean age of all unsurvived & missing age of each survived by mean age of all survived.
```

```python
import math
sumAgeSurvived,sumAgeUnsurvive,countSurvived,countUnsurvived=0,0,0,0
for i in range(0,891,1):
    if not(math.isnan(titanic.at[i,'Age'])):
        if titanic.at[i,'Survived']==1:
            sumAgeSurvived+=titanic.at[i,'Age']
            countSurvived+=1
        else:
            sumAgeUnsurvive+=titanic.at[i,'Age']
            countUnsurvived+=1
    else:
        continue
        
meanSurvivedAge=sumAgeSurvived/countSurvived
meanUnsurvivedAge=sumAgeUnsurvive/countUnsurvived

for i in range(0,891,1):
    if math.isnan(titanic.at[i,'Age']):
        if titanic.at[i,'Survived']==1:
            titanic.at[i,'Age']=meanSurvivedAge
        else:
            titanic.at[i,'Age']=meanUnsurvivedAge
    else:
        continue
```
