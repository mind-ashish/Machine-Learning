```python
import pandas as pd
a=pd.read_csv("annual-enterprise-survey-2017-financial-year-provisional-csv.csv")
a
```

```python
###Data cleaning. Dropping unwanted rows & cols:
a.drop(0)
#show data after drop row 0 , a remains same.
```

```python
a
```

```python
a.drop(0)
```

```python
a.drop(0, inplace=True)  #actually drops row 0 of a, a will change
```

```python
a
```

```python
#to drop a row by its name(not deflected in a, to change a use , inplace =True) 
a.drop(3)  #here 3 is name of row not index
```

```python
# to drop a column by name
a.drop('Year', axis=1)  #year is column name, axis=1 means we are dropping column.
```

```python
#to drop multiple row by thier name, pass list of row names.
a.drop([2,3,5])   #2,3,5 are row names
```

```python
#to drop multiple columns by name
a.drop(["Year","Units","Value"],axis=1)   #Year,Units,Value are col name, axis =1 means columns.
```

```python
#dropping rows by passing range
a.drop([2,5])  #drops row starting from name=2 to 5 ,i.e 2,3,4     #we mention start index & end index by that row name.
```

```python
#drop columns by passing index range
a.drop(["Year","Units"],axis=1)   #we mention start index & end index by that col name.
```

```python
b=a.copy()  #to make copy of a ,any change in b willnot reflect in a.
```
