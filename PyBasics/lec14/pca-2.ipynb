```python
#here we are trying to covert 3d data into 2d data by using pca. i.e 3 dimesnions to 
#2 dimesnions.
import numpy as np
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
#we want a 3d data, so three co-ordinates of each pt. And 2 classes which data belong to.
np.random.seed(2343242)     #to make generate dummy data points randomly using numpy
```

```python
mean_vec1=np.array([0,0,0])    #mean of pts on class1
cov_mat1=np.array([[1,0,0],[0,1,0],[0,0,1]])   #covariance
class1=np.random.multivariate_normal(mean_vec1,cov_mat1,100)  #this fn need mean,covariance, and no of 
#pts i.e 100-> 100 data pts here.
```

```python
#similarly can make pts of other class
mean_vec2=np.array([1,1,1])    #just take diff mean
cov_mat2=np.array([[1,0,0],[0,1,0],[0,0,1]])
class2 =np.random.multivariate_normal(mean_vec2,cov_mat2,100)
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D,proj3d   #to plot 3d graph
```

```python
fig=plt.figure(figsize=(8,8))     #it makes 8 by 8 inches figure plot
ax=fig.add_subplot(111,projection='3d')     #111 means fig will hv ,1 row, 1column and we work on 1th entry
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
plt.show()
```

```python
#to move this 3d to 2d
```

```python
combined_data=np.concatenate((class1,class2)) #to combine data of both classes.
```

```python
combined_data
```

```python
pca=PCA()      # 3d to 3d
```

```python
transformed_data=pca.fit_transform(combined_data)
```

```python
transformed_data   #still has 3 dimensions
```

```python
#for 2 dimesnions
```

```python
pca=PCA(n_components=2)
```

```python
transformed_data=pca.fit_transform(combined_data)
```

```python
transformed_data  #we kept 2 imp dimensions/line only
```

```python
pca.components_
```

```python
# if we plot this 2d data
```

```python
plt.scatter(transformed_data[:,0],transformed_data[:,1])
plt.show()
```

```python
#plotting the transformed data, we get a 2d plane, since first 100 rows-> class1 marking with "o" ,and next 100->class 2
plt.plot(transformed_data[0:100,0],transformed_data[0:100,1],"o")
plt.plot(transformed_data[100:200,0],transformed_data[100:200,1],"^")
plt.show()
```

```python
#if we inverse transform 2d to 3d , it will be a all points liying on same plane,
#just like all points were linear when we did 1d to 2d inverse transform.
X_approx=pca.inverse_transform(transformed_data)
X_approx   #3d again 
```

```python
fig=plt.figure(figsize=(8,8))     #it makes 8 by 8 inches figure plot
ax=fig.add_subplot(111,projection='3d')     #111 means fig will hv ,1 row, 1column and we work on 1th entry
ax.plot(X_approx[:,0],X_approx[:,1],X_approx[:,2],"*")
plt.show()
```
